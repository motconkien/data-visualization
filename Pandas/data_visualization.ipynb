# Data visualization using pandas plot directly

```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('data/employees_full.csv')
df1 = df.head()
```

## Bar plot

```python
df1[['SALARY','MANAGER_ID','LAST_NAME']].plot.bar()
```

```python
# Add x title and y title, bar title
df1[['SALARY','MANAGER_ID','LAST_NAME']].plot.bar(title = 'Salary of Top Manager', x = 'LAST_NAME', y = 'SALARY', rot = 60)
```

```python
# Stastitic number of employees by Department
df2 = df.DEPARTMENT_NAME.value_counts()
df2.plot.bar(title = 'Number employees of Department', rot = 75)
```

```python
# Calculate the min, max and mean salary group by Department and sort decending mean value then taking only first 5 values
sal_df = df.groupby('DEPARTMENT_NAME')['SALARY'].agg(['min', 'max','mean']).sort_values('mean', ascending=False).head()
sal_df
```

```python
sal_df[['min', 'max']].plot.bar(figsize = (10,6), rot = 0, title = 'Min and Max salary of Department')
```

```python
#Statistics on the number employees by year and margin = True to ensure to display totals of the rows and columns
df_cnt = pd.crosstab(index=df['DEPARTMENT_NAME'], columns=df['HIRE_YEAR'], margins=True)
df_cnt
```

```python
df_cnt = df_cnt.sort_values(by ='All').iloc[:-1,:-1] #remove the all columns all as well
```

```python
df_cnt.plot.barh(stacked = True, title = 'Statistics on the number employees by year')
```

## Line Plot

```python
df_cnt = df.HIRE_YEAR.value_counts().reset_index().rename(columns={'count': 'COUNT'}).sort_values(by = 'HIRE_YEAR')
df_cnt
```

```python
df_cnt.plot(x = 'HIRE_YEAR', title = 'Number employees each year')
```

```python
sal_df.plot(y = 'min', title = 'Min salary of Department')
```
