```python
import numpy as np 
import pandas as pd 
```

```python
#reading file
emp_df = pd.read_csv('data/employees_astyle.csv')
emp_df.head()
```

```python
emp_df.dtypes
```

```python
#Add new col is 'HIRE_YEAR' and change 'HIRE_DATE' to datetime type
emp_df['HIRE_YEAR']= pd.to_datetime(emp_df['HIRE_DATE']).dt.year
emp_df['HIRE_DATE']= pd.to_datetime(emp_df['HIRE_DATE'])
emp_df
```

```python
#Reading more file
dep_df = pd.read_excel('data/departments.xls', engine='xlrd')
dep_df.head()
```

```python
#Merge 2 dataframe
#Note: when merge, need to import the key column
emp_df = pd.merge(emp_df,dep_df[['DEPARTMENT_ID', 'DEPARTMENT_NAME']])
emp_df
```

From the summary statistic, the information can be taken here is:
* count: the number of non null entries
* mean: the mean value
* std: the standard deviation
* min: the minimum value
* 25%: The 25th percentile (first quartile)
* 50%: The 50th percentile (median)
* 75%: The 75th percentile (third quartile)
* max: the max value
* unique: the number of unique value 
* top: the frequency value
* freq: frequency of the most frequent value

## CROSSTAB and PIVOT TABLE, MELT

```python
pd.crosstab(index=emp_df['HIRE_YEAR'], columns=emp_df['DEPARTMENT_NAME'])
```

```python
# Calculate sum salary by department and fill null value = 0
pivot = pd.pivot_table(data=emp_df,columns='DEPARTMENT_NAME', index='HIRE_YEAR', values='SALARY', aggfunc='sum', fill_value=0)
pivot
```

```python
#unpivot table by using melt
unpivot = pd.melt(frame= pivot.reset_index(),id_vars='HIRE_YEAR',value_vars=pivot.columns,var_name='DEPARTMENT',value_name='SALARY')
unpivot
```
